```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib inline

# 指定运行的GPU 
os.environ["CUDA_VISIBLE_DEVICES"]="1"
```

```python
# 读取数据
# PassengerId 整型变量，标识乘客的ID，递增变量，对预测无帮助
# Survived 整型变量，标识该乘客是否幸存。0表示遇难，1表示幸存
# Pclass 整型变量，标识乘客的社会-经济状态，1代表Upper，2代表Middle，3代表Lower
# Name 字符型变量，除包含姓和名以外，还包含Mr. Mrs. Dr.这样的具有西方文化特点的信息
# Sex 字符型变量，标识乘客性别，适合转换为factor类型变量
# Age 整型变量，标识乘客年龄，有缺失值
# SibSp 整型变量，代表兄弟姐妹及配偶的个数。其中Sib代表Sibling也即兄弟姐妹，Sp代表Spouse也即配偶
# Parch 整型变量，代表父母或子女的个数。其中Par代表Parent也即父母，Ch代表Child也即子女
# Ticket 字符型变量，代表乘客的船票号
# Fare 数值型，代表乘客的船票价
# Cabin 字符型，代表乘客所在的舱位，有缺失值
# Embarked 字符型，代表乘客登船口岸
train_data = pd.read_csv("train.csv")
train_data = train_data.drop(columns=['PassengerId','Name','Ticket','Cabin','Embarked','SibSp','Parch'])
train_data.isnull().sum()
```

```python
# 逐个特征单独分析，确定相关性
# Pclass
Survived = train_data['Survived']
x_label = [1,2,3]
Pclass_0 = train_data[train_data['Survived']==0]['Pclass']
Pclass_1 = train_data[train_data['Survived']==1]['Pclass']
enumerat_0 = pd.value_counts(Pclass_0).sort_index(ascending=False)
enumerat_1 = pd.value_counts(Pclass_1).sort_index(ascending=False)
total_width, n = 0.8, 2
width = total_width / n
x = list(range(len(x_label)))
m = plt.bar(x,enumerat_0, label='0',width=width)
for i in range(len(x)):
    x[i] += width
n = plt.bar(x,enumerat_1, label='1', tick_label=x_label,width=width)

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        plt.text(rect.get_x()+rect.get_width()/2.-0.1, 1.01*height, '%s' % int(height))

autolabel(m)
autolabel(n)
    
plt.legend()
```

```python
def iv_woe(data:pd.DataFrame, target:str, bins:int = 10) -> (pd.DataFrame, pd.DataFrame):
    """计算woe和IV值
    
    参数：
    - data: dataframe数据
    - target: y列的名称
    - bins: 分箱数（默认是10）
    """
    newDF,woeDF = pd.DataFrame(), pd.DataFrame()
    cols = data.columns
    for ivars in cols[~cols.isin([target])]:
        # 数据类型在bifc中、且数据>10则分箱
        if (data[ivars].dtype.kind in 'bifc') and (len(np.unique(data[ivars]))>10):
            binned_x = pd.qcut(data[ivars], bins,  duplicates='drop')
            d0 = pd.DataFrame({'x': binned_x, 'y': data[target]})
        else:
            d0 = pd.DataFrame({'x': data[ivars], 'y': data[target]})
        d = d0.groupby("x", as_index=False).agg({"y": ["count", "sum"]})
        d.columns = ['Cutoff', 'N', 'Events']
        d['% of Events'] = np.maximum(d['Events'], 0.5) / d['Events'].sum()
        d['Non-Events'] = d['N'] - d['Events']
        d['% of Non-Events'] = np.maximum(d['Non-Events'], 0.5) / d['Non-Events'].sum()
        d['WoE'] = np.log(d['% of Events']/d['% of Non-Events'])
        d['IV'] = d['WoE'] * (d['% of Events'] - d['% of Non-Events'])
        d.insert(loc=0, column='Variable', value=ivars)
        print("Information value of " + ivars + " is " + str(round(d['IV'].sum(),6)))
        temp =pd.DataFrame({"Variable" : [ivars], "IV" : [d['IV'].sum()]}, columns = ["Variable", "IV"])
        newDF=pd.concat([newDF,temp], axis=0)
        woeDF=pd.concat([woeDF,d], axis=0)
    return newDF, woeDF
```

```python
# 清洗数据
train_data.describe(include='all')
```

```python
# 取值映射，转换成数值 sex，embarked
sex_mapping = {'male':0,'female':1}
train_data['Sex'] = train_data['Sex'].replace(sex_mapping)
train_data.describe(include='all')
```

```python
# 缺失值处理
train_data.isnull().sum()
```

```python
# 平均数补充Age的缺失值,方案待定
train_data['Age'] = train_data['Age'].fillna(np.mean(train_data['Age']))
train_data.isnull().sum()
```

```python
train_target = train_data['Survived']
train_data = train_data.drop(columns=['Survived'])
train_data.describe(include='all')
```

```python
# 数据归一
max_min_scaler = lambda x : (x-np.min(x))/(np.max(x)-np.min(x))
train_data['Fare'] = train_data[['Fare']].apply(max_min_scaler)
train_data['Age'] = train_data[['Age']].apply(max_min_scaler)
train_data.describe(include='all')
```

```python
# 回调函数
from keras import callbacks
path_checkpoint = 'sentiment_checkpoint.keras'
checkpoint = callbacks.ModelCheckpoint(filepath=path_checkpoint, monitor='loss',verbose=1, save_weights_only=True, save_best_only=True)
earlystopping = callbacks.EarlyStopping(monitor='loss', patience=5, verbose=1)
lr_reduction = callbacks.ReduceLROnPlateau(monitor='loss',factor=0.1, min_lr=1e-8, patience=0,verbose=1)
callbacks=[earlystopping,checkpoint,lr_reduction]
```

```python
# 基准模型
from keras import models
from keras import layers
from keras.optimizers import Adam

def create_model(layer_num=2,units=40,optimizer='adam',activation='relu'):
    model = models.Sequential()
    model.add(layers.Dense(units=units,input_dim=4,activation=activation))
    for i in range(layer_num):
        model.add(layers.Dense(units=units,activation='relu'))
    model.add(layers.Dense(units=1,activation='sigmoid'))
    model.compile(loss='binary_crossentropy',optimizer=optimizer,metrics=['accuracy'])
    return model
```

```python
# 超参数优化
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV

seed = 5
np.random.seed(seed)
estimator = KerasClassifier(build_fn=create_model, nb_epoch=20, batch_size=5, verbose=0)

kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)

optimizer = ['SGD', 'RMSprop', 'Adagrad', 'Adadelta', 'Adam', 'Adamax', 'Nadam']
activation = ['relu','tanh','sigmoid','softmax']
layer_num = [0,1,2,3,4]
units = [10,20,30,40,50,60,70,80]
param_grid = dict(layer_num=layer_num,units=units,optimizer=optimizer,activation=activation)
grid = GridSearchCV(estimator=estimator, param_grid=param_grid,cv=kfold, verbose=1)
grid_result = grid.fit(train_data, train_target)

# 打印结果
print('Best: {} using {}'.format(grid_result.best_score_, grid_result.best_params_))
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']

for mean, std, param in zip(means, stds, params):
    print("%f (%f) with: %r" % (mean, std, param))
```

```python
# 交叉验证
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score


seed = 5
np.random.seed(seed)

estimator = KerasClassifier(build_fn=create_model, nb_epoch=20, batch_size=5, verbose=0)
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)
results = cross_val_score(estimator, train_data, train_target, cv=kfold)
print("Baseline: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```
